```python
import pandas as pd
import numpy as np

LINE_1 = []
LINE_2 = []
LINE_3 = []
LINE_4 = []

LINE_5 = []
LINE_6 = []

for i in np.arange(1, 101, 1):
    LINE_1.append(i)
    LINE_2.append(0.5 * i)
    LINE_3.append((1000) - i)
    LINE_4.append((0.8 * (i**2)) + (5*i) + 200)

for i in np.arange(-100, 101, 1):
    LINE_5.append(i)
    LINE_6.append((0.5 * (i**3)) - (i**2) + (0.1 * i) + 20)

df = pd.DataFrame({'X': LINE_1, 'Y1': LINE_2, 'Y2': LINE_3, 'Y3': LINE_4})

df_2 = pd.DataFrame({'X': LINE_5, 'Y1': LINE_6})
```

```python
import matplotlib.pyplot as plt

plt.plot(df['X'], df['Y1'])
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Regresion Lineal')
plt.savefig('Images/regre_exam_1.png')
```

```python
plt.plot(df['X'], df['Y2'])
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Regresion Inversa')
plt.savefig('Images/regre_exam_2.png')
```

```python
plt.plot(df['X'], df['Y3'])
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Regresion Polinomica de Grado 2')
plt.savefig('Images/regre_exam_3.png')
```

```python
plt.plot(df_2['X'], df_2['Y1'])
plt.xlabel('Variable Independiente')
plt.ylabel('Variable Dependiente')
plt.title('Regresion Polinomica de Grado 3')
plt.savefig('Images/regre_exam_4.png')
```
